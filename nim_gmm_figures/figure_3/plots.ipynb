```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Load the saved clusters from the specified iterations
initial_clusters = np.load('cluster_outputs/initial_clusters.npy')
clusters_iter3 = np.load('cluster_outputs/clusters_after_iteration_3.npy')
clusters_iter5 = np.load('cluster_outputs/clusters_after_iteration_5.npy')
clusters_iter7 = np.load('cluster_outputs/clusters_after_iteration_7.npy')

# Extract coordinates and labels
xyz_initial = initial_clusters[:, :3]
labels_initial = initial_clusters[:, 3]

xyz_iter3 = clusters_iter3[:, :3]
labels_iter3 = clusters_iter3[:, 3]

xyz_iter5 = clusters_iter5[:, :3]
labels_iter5 = clusters_iter5[:, 3]

xyz_iter7 = clusters_iter7[:, :3]
labels_iter7 = clusters_iter7[:, 3]

# Set plot limits and grid size
x_limits = (124, 140)
y_limits = (120, 190)
pixel_size_mm = 2

# Create a 2x2 subplot for double-column format
fig, axs = plt.subplots(2, 2, figsize=(12, 8), dpi=300)

# Function to add rectangle patches for each point
def add_rectangles(ax, xyz_data, labels, cmap):
    for (x, y, _), label in zip(xyz_data, labels):
        color = cmap(int(label) % cmap.N)
        rect = patches.Rectangle((x, y), pixel_size_mm, pixel_size_mm, linewidth=0.5,
                                 edgecolor='none', facecolor=color, alpha=0.7)
        ax.add_patch(rect)

# Function to set custom grid and sparse Y-axis tick labels
def set_custom_grid(ax):
    ax.set_xlim(x_limits)
    ax.set_ylim(y_limits)
    
    # Set grid ticks every 2 mm on both axes
    ax.set_xticks(np.arange(x_limits[0], x_limits[1] + 1, pixel_size_mm), minor=False)
    ax.set_yticks(np.arange(y_limits[0], y_limits[1] + 1, pixel_size_mm), minor=True)
    ax.set_xticks(np.arange(x_limits[0], x_limits[1] + 1, pixel_size_mm), minor=True)
    ax.set_yticks(np.arange(y_limits[0], y_limits[1] + 1, 8), minor=False)
    
    # Display grid
    ax.grid(which="both", color="gray", linestyle="--", linewidth=0.5)

    # Label Y-axis major ticks every 8 mm
    y_labels = np.arange(y_limits[0], y_limits[1] + 1, 8)
    ax.set_yticklabels(y_labels)
    
# Define a colormap
cmap = plt.cm.get_cmap("viridis", len(np.unique(labels_initial)))

# Plot Initial Clusters
set_custom_grid(axs[0, 0])
add_rectangles(axs[0, 0], xyz_initial, labels_initial, cmap)
axs[0, 0].set_xlabel('X [mm]', fontname='Arial', fontsize=10)
axs[0, 0].set_ylabel('Y [mm]', fontname='Arial', fontsize=10)
axs[0, 0].text(0.05, 0.95, '(a): Iteration-1', transform=axs[0, 0].transAxes, fontsize=14, verticalalignment='top')

# Plot Clusters after Iteration 3
set_custom_grid(axs[0, 1])
add_rectangles(axs[0, 1], xyz_iter3, labels_iter3, cmap)
axs[0, 1].set_xlabel('X [mm]', fontname='Arial', fontsize=10)
axs[0, 1].set_ylabel('Y [mm]', fontname='Arial', fontsize=10)
axs[0, 1].text(0.05, 0.95, '(b): Iteration-3', transform=axs[0, 1].transAxes, fontsize=14, verticalalignment='top')

# Plot Clusters after Iteration 5
set_custom_grid(axs[1, 0])
add_rectangles(axs[1, 0], xyz_iter5, labels_iter5, cmap)
axs[1, 0].set_xlabel('X [mm]', fontname='Arial', fontsize=10)
axs[1, 0].set_ylabel('Y [mm]', fontname='Arial', fontsize=10)
axs[1, 0].text(0.05, 0.95, '(c): Iteration-5', transform=axs[1, 0].transAxes, fontsize=14, verticalalignment='top')

# Plot Clusters after Iteration 7
set_custom_grid(axs[1, 1])
add_rectangles(axs[1, 1], xyz_iter7, labels_iter7, cmap)
axs[1, 1].set_xlabel('X [mm]', fontname='Arial', fontsize=10)
axs[1, 1].set_ylabel('Y [mm]', fontname='Arial', fontsize=10)
axs[1, 1].text(0.05, 0.95, '(d): Iteration-7', transform=axs[1, 1].transAxes, fontsize=14, verticalalignment='top')

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.savefig('cluster_visualization_with_rectangles.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Load the saved clusters from the specified iterations
initial_clusters = np.load('cluster_outputs/initial_clusters.npy')
clusters_iter3 = np.load('cluster_outputs/clusters_after_iteration_3.npy')
clusters_iter5 = np.load('cluster_outputs/clusters_after_iteration_5.npy')
clusters_iter7 = np.load('cluster_outputs/clusters_after_iteration_7.npy')

# Extract coordinates and labels
xyz_initial = initial_clusters[:, :3]
labels_initial = initial_clusters[:, 3]

xyz_iter3 = clusters_iter3[:, :3]
labels_iter3 = clusters_iter3[:, 3]

xyz_iter5 = clusters_iter5[:, :3]
labels_iter5 = clusters_iter5[:, 3]

xyz_iter7 = clusters_iter7[:, :3]
labels_iter7 = clusters_iter7[:, 3]

# Set plot limits and grid size
x_limits = (124-50, 140+50)
y_limits = (120, 190)
pixel_size_mm = 2

# Create a 2x2 subplot for double-column format
fig, axs = plt.subplots(2, 2, figsize=(12, 8), dpi=300, constrained_layout=True)

# Function to add rectangle patches for each point
def add_rectangles(ax, xyz_data, labels, cmap):
    for (x, y, _), label in zip(xyz_data, labels):
        color = cmap(int(label) % cmap.N)
        rect = patches.Rectangle((x, y), pixel_size_mm, pixel_size_mm, linewidth=0.5,
                                 edgecolor='none', facecolor=color, alpha=0.7)
        ax.add_patch(rect)

# Function to set custom grid and sparse Y-axis tick labels
def set_custom_grid(ax):
    ax.set_xlim(x_limits)
    ax.set_ylim(y_limits)
    
    # Set grid ticks every 2 mm on both axes
    ax.set_xticks(np.arange(x_limits[0], x_limits[1] + 1, 8), minor=False)
    ax.set_yticks(np.arange(y_limits[0], y_limits[1] + 1, pixel_size_mm), minor=True)
    ax.set_xticks(np.arange(x_limits[0], x_limits[1] + 1, pixel_size_mm), minor=True)
    ax.set_yticks(np.arange(y_limits[0], y_limits[1] + 1, 8), minor=False)
    
    # Display grid
    ax.grid(which="both", color="lightgray", linestyle="--", linewidth=0.5)

    # Label Y-axis major ticks every 8 mm
    y_labels = np.arange(y_limits[0], y_limits[1] + 1, 8)
    ax.set_yticklabels(y_labels)

# Define a colormap
cmap = plt.cm.get_cmap("viridis", len(np.unique(labels_initial)))

# Plot Initial Clusters
set_custom_grid(axs[0, 0])
add_rectangles(axs[0, 0], xyz_initial, labels_initial, cmap)
axs[0, 0].set_xlabel('X [mm]', fontname='Arial', fontsize=12, fontweight='bold')
axs[0, 0].set_ylabel('Y [mm]', fontname='Arial', fontsize=12, fontweight='bold')
axs[0, 0].text(0.05, 0.95, '(a): Iteration-1', transform=axs[0, 0].transAxes, fontsize=14, verticalalignment='top', fontweight='bold')

# Plot Clusters after Iteration 3
set_custom_grid(axs[0, 1])
add_rectangles(axs[0, 1], xyz_iter3, labels_iter3, cmap)
axs[0, 1].set_xlabel('X [mm]', fontname='Arial', fontsize=12, fontweight='bold')
axs[0, 1].set_ylabel('Y [mm]', fontname='Arial', fontsize=12, fontweight='bold')
axs[0, 1].text(0.05, 0.95, '(b): Iteration-3', transform=axs[0, 1].transAxes, fontsize=14, verticalalignment='top', fontweight='bold')

# Plot Clusters after Iteration 5
set_custom_grid(axs[1, 0])
add_rectangles(axs[1, 0], xyz_iter5, labels_iter5, cmap)
axs[1, 0].set_xlabel('X [mm]', fontname='Arial', fontsize=12, fontweight='bold')
axs[1, 0].set_ylabel('Y [mm]', fontname='Arial', fontsize=12, fontweight='bold')
axs[1, 0].text(0.05, 0.95, '(c): Iteration-5', transform=axs[1, 0].transAxes, fontsize=14, verticalalignment='top', fontweight='bold')

# Plot Clusters after Iteration 7
set_custom_grid(axs[1, 1])
add_rectangles(axs[1, 1], xyz_iter7, labels_iter7, cmap)
axs[1, 1].set_xlabel('X [mm]', fontname='Arial', fontsize=12, fontweight='bold')
axs[1, 1].set_ylabel('Y [mm]', fontname='Arial', fontsize=12, fontweight='bold')
axs[1, 1].text(0.05, 0.95, '(d): Iteration-7', transform=axs[1, 1].transAxes, fontsize=14, verticalalignment='top', fontweight='bold')

# Make the border thicker
for ax in axs.flat:
    ax.set_aspect('equal', adjustable='box')
    ax.spines['top'].set_linewidth(1.5)
    ax.spines['right'].set_linewidth(1.5)
    ax.spines['bottom'].set_linewidth(1.5)
    ax.spines['left'].set_linewidth(1.5)

fig.subplots_adjust(wspace=0.05, hspace=0.1)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.savefig('cluster_visualization_with_rectangles.png', dpi=300, bbox_inches='tight')
plt.show()
```
