```python
import numpy as np
import matplotlib.pyplot as plt

# Load the data
data = np.load('sim_10mev_3cm_event335.npy')

x = data[:, 0]
y = data[:, 1]

# Cluster labels
gmm_labels = data[:, 8]
db_labels  = data[:, 9]
pkl_labels = data[:, 10]
ckl_labels = data[:, 11]

label_dict = {
    "DB Clusters": db_labels,
    "GMM Clusters": gmm_labels,
    "PKL Labels": pkl_labels,
    "CKL Labels": ckl_labels
}

# Axis limits
xlim = [x.min(), x.max()]
ylim = [y.min(), y.max()]

# Data aspect ratio (Y/X)
data_aspect = (ylim[1] - ylim[0]) / (xlim[1] - xlim[0])

# Subplot grid
nrows, ncols = 2, 2
subplot_width = 4
fig_width = ncols * subplot_width
fig_height = nrows * subplot_width * data_aspect

fig, axs = plt.subplots(nrows, ncols, figsize=(fig_width, fig_height))
axs = axs.flatten()

for idx, (ax, labels) in enumerate(zip(axs, label_dict.values())):
    ax.scatter(x, y, c=labels, s=5, cmap="tab20", alpha=0.7)
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
    ax.set_aspect('equal', 'box')
    
    # Only show x-axis labels for bottom row
    if idx < ncols:  # top row
        ax.set_xticklabels([])
    else:  # bottom row
        ax.set_xlabel("X", fontsize=8)
        ax.tick_params(axis='x', labelsize=8)  # no rotation
    
    # Only show y-axis labels for left column
    if idx % ncols != 0:  # right column
        ax.set_yticklabels([])
    else:  # left column
        ax.set_ylabel("Y", fontsize=8)
        ax.tick_params(axis='y', labelsize=8)

plt.subplots_adjust(wspace=0.05, hspace=0.05)
plt.show()
```
